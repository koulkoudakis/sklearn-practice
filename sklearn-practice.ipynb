### Option 2: Missing values with sklearn

```python
car_sales_missing = pd.read_csv("data/car-sales-extended-missing-data.csv")
car_sales_missing.head(10)
```

```python
car_sales_missing.isna().sum()
```

```python
# Drop rows with no labels
car_sales_missing.dropna(subset=["Price"], inplace=True)
car_sales_missing.isna().sum()
```

```python
# Split into X and y
X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]
```

```python
# Fill missing values with Scikit-Learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split



# Fill categorical values with 'missing' and numerical values with mean
cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy="constant", fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

# Define columns
cat_features = ["Make", "Colour"]
door_feature = ["Doors"]
num_features = ["Odometer (KM)"]

# Create imputer (to fill missing data)
imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_features),
    ("door_imputer", door_imputer, door_feature),
    ("num_imputer", num_imputer, num_features)
])

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Transform data
filled_X_train = imputer.fit_transform(X_train)
filled_X_test = imputer.fit_transform(X_test)

# Preview data
filled_X_train
```

```python
len(filled_X_train), len(filled_X_test)
```

```python
car_sales_filled_train = pd.DataFrame(filled_X_train,
                                     columns=["Make", "Colour", "Doors", "Odometer (KM)"]
                                     )
car_sales_filled_test = pd.DataFrame(filled_X_test,
                                     columns=["Make", "Colour", "Doors", "Odometer (KM)"]
                                     )
```

```python
# No missing values
car_sales_filled_train.isna().sum(), car_sales_filled_test.isna().sum()
```

```python
# Turn categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                                remainder="passthrough")
transformed_X_train = transformer.fit_transform(car_sales_filled_train)
transformed_X_train
```

```python
transformed_X_test = transformer.fit_transform(car_sales_filled_test)
transformed_X_test
```

```python
# Data is transformed to numbers and filled
# Fitting model
np.random.seed(18)
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(transformed_X_train, y_train)
model.score(transformed_X_test, y_test)
```

### 2.1 Picking a machine learning model for a regression problem

```python
# Import Bosting housing dataset
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
type(boston)
```

```python
# Number of samples
len(boston["data"])
```

```python
boston_df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
boston_df["target"] = pd.Series(boston["target"])
boston_df.head()
```

```python
# Using Ridge regression model
from sklearn.linear_model import Ridge

# Setup random seed
np.random.seed(18)

# Create data
X = boston_df.drop("target", axis=1)
y = boston_df["target"]

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiate Ridge model
model = Ridge()
model.fit(X_train, y_train)

# Check score of Ridge model on test data
model.score(X_test, y_test)
```


### Improving score...

```python
# Trying random forest regressor
from sklearn.ensemble import RandomForestRegressor

# Set up random seed
np.random.seed(18)

# Create data
X = boston_df.drop("target", axis=1)
y = boston_df["target"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiate Random Forest Regressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

# Evaluate Random Forest Regressor
rf.score(X_test, y_test)
```

### 2.2 Choosing an estimator for classification problem

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease.head()
```

`LinearSVC` seems like the best option for our dataset after consulting the sk-learn map

```python
# Import the LinearSVC estimator class
from sklearn.svm import LinearSVC

# Setup random seed
np.random.seed(18)

# Create data
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiate LinearSVC
clf = LinearSVC(max_iter=10000)
clf.fit(X_train, y_train)

# Evaluate LinearSVC
clf.score(X_test, y_test)
```

```python
# Import the RandomForestClassifier estimator class
from sklearn.ensemble import RandomForestClassifier

# Setup random seed
np.random.seed(18)

# Create data
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiate RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Evaluate RandomForestClassifier
clf.score(X_test, y_test)
```
